#Matplotlib

Function : It is used to work with graphs and charts

Importing Command : import matplotlib.pyplot as plt

Sub-Functions : 1.plt.plot (For Plotting the graph)
                2.plt.figure (For Showing and changing the figure Size)
                3.plt.show  (For Showing the graph )
                4.plt.xlabel , plt.ylabel (For labelling the axis of the graph )
                5.plt.title (For adding a title of graph on top )
                6.plt.tick_params (For adding parameters to values on axis )
                7.plt.subplot  (To Plot multiple graphs at same time )
                8.plt.suptitle (To add super titles to subplotted graphs )
                9.plt.bar  (To Plot bar graph )
                10.plt.hist  (To Plot Histogram )
                11.plt.scatter (To Plot Scatter Graph )
                12.plt.legend (For Adding Legend to Graph )
                13.ax.annotate (To Add Annotations to the Graph )

# Graph plot using matplotlib 

```python
#import matplotlib
import matplotlib.pyplot as plt

#x-axis co-ordinates
x= [1,2,3]

#y-axis co-ordinates
y= [4,5,6]

#plotting the graph
plt.plot(x,y)

#Showing figure size
fig= plt.figure()

#showing the graph
plt.show()
```

# Showing colors and markers

```python
import numpy as np
import matplotlib.pyplot as plt

x= np.array([2,3,8,7])
y= np.array([1,4,1,9])

plt.plot(x,y, marker= '+',color='g',linestyle='dotted')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Line Plot")
plt.show()
```

```python
import random as ran

students =["Aa","Bb","Cc","Dd","Ee"]

marks=[]
for i in range(0,len(students)):marks.append(ran.randint(0,101))

plt.xlabel("Students")
plt.xticks(rotation = 45)
plt.ylabel("Marks")
plt.title("CLASS RECORDS")
plt.plot(students,marks,'g--')
```

```python
# Ticks are the markers denoting data points an axes
# importing libraries

import matplotlib.pyplot as plt
import numpy as np

# values of x and y axes

x = [5,10,15,20,25,30,35,40,45,50]
y = [1,4,3,2,7,6,9,8,10,5]

plt.plot(x,y,'g')
plt.xlabel('x')
plt.ylabel('y')

# here we set the size for ticks, rotation and color value

plt.tick_params(axis='x', labelsize = 18, labelrotation = 90, labelcolor = 'blue')
plt.tick_params(axis='y', labelsize = 12, labelrotation = 190, labelcolor = 'black')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#draw plot as fig
fig, ax = plt.subplots()


x = np.linspace(0,20,1000)
ax.plot(x, np.cos(x))
ax.axis('equal')

#text annotation to the plot where it indicate maximum value of the
ax.annotate('local maximum', xy=(6.28,1),xytext=(10,4),arrowprops=dict(facecolor='black', shrink =0.05))

#text annotation to the plot where it indicate minimum value of the
ax.annotate('local minimum', xy=(5* np.pi,-1) ,xytext =(2,-6),arrowprops=dict(arrowstyle ="->", connectionstyle ="angle3,angleA=0,angleB=90"))
```

# Adding annotate & legend

```python
import numpy as np
import matplotlib.pyplot as plt
#You can also write    
#from matplotlib import pyplot as plt


x = np.arange(0,10)
y = x^2
z = x^3

plt.plot(x,y)
plt.plot(x,z)
plt.plot(y,z)

plt.xlabel('Time')
plt.ylabel('Distance')

plt.title('My First Plot')

             
# Annotation

plt.annotate(xy = [2,1], text='first entry', rotation = 20)
#plt.annotate(xy = [2,1], text = "First Entry")
#plt.annotate(xy = [4,6], text = "Second Entry")
#plt.annotate(xy = [8,9], text = "Third Entry")

plt.legend(["race 1","race 2","race 3"], loc = 4)

plt.show()
```

#  Pie plot

pie chart is a circular statistical plot that can display only one series of data 

the area of the chart is the total % of the given data

```python
#import libraries

import numpy as np
import matplotlib.pyplot as plt

#Creating a dataset

Branch = ["Computer","Electrical","Mechanical","Civil","Automobile","IT"]

data = [50,34,40,29,12,23]
explode = (0.2,0,0,0,0.3,0,)

#Create chart 

fig = plt.figure(figsize=(10,7))
plt.pie(data, labels=Branch, explode=explode, autopct = '%1.1f%%',shadow=True)

plt.show()
```

#  3D plotting

```python
#import mplt3d toolkit,numpy & matplotlib

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d #3d toolkit lib

```

```python
fig = plt.figure()

#syntax for 3D projection

ax = plt.axes(projection = '3d')

#Defining all 3 axes 

z = np.linspace(0,1,100)
x = z*np.cos(25*z)
y = z*np.sin(25*z)

#3d plotting

ax.plot3D(x,y,z,'green')
ax.set_title('3D plotting')
plt.show()
```
